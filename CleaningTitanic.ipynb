```python
import pandas as pd

# Els fitxers han sigut trets de la Wikipedia en format Excel:
tit=pd.read_excel('all_Titanic.xlsx')
tit.head()
```

```python
# Afegim, per conveniència de l'anàlisi, una columna de salvats
# Els salvats van ser els que pujaren a un bot:
tit['Survived']= ~tit['Lifeboat'].isna()
tit.head()
```

```python
# Name podria ser un identificador, si no haguessin duplicats de nom:
tit[tit.Name.duplicated()]
```

```python
# Hi ha noms duplicats legítims, comprovats a la Wikipedia, amb diferent edat, o càrrec, o
# u s'ha salvat, l'altre, no, o són de diferent classe, o de diferent grup;
# a soles el cuiner Auguste Louis Coutin està duplicat: treballa en Victualling i Restaurant.
# La resta de les seves dades són idèntiques.
tit[tit.Name=='Coutin, Mr. Auguste Louis']
```

```python
# Eliminem la línia 2043 del cuiner Auguste Louis Coutin i deixem la 1730:
tit.drop([2043], axis=0, inplace=True)
tit[tit.Name=='Coutin, Mr. Auguste Louis']
```

```python
# El nom no és un bon identificador; per tant cal afegir un identificador ID:
tit['ID']=tit.index
tit.head()
```

```python
# Comprovem els salvats en la nova columna Survived
tit.Survived.sum()
```

```python
# Comprovem els NA
tit.isna().sum()
```

```python
# Comprovem primer el NA d'Age
tit[tit.Age.isna()]
```

```python
# Calculem l'edat mediana de cada càrrec
position_age=tit.groupby('Position').Age.median()
print(position_age)
```

```python
# La mediana d'edat dels 'Trimmers' d'enginnering:
position_age['Trimmer']
```

```python
# Assignem a Mr. A. Dore la mediana d'edat dels 'Trimmers' d'enginnering
tit.loc[[1401],['Age']]=position_age['Trimmer']
```

```python
# Comprovem que està rectificat:
tit.loc[1401]
```
